Given a string array words, return the maximum value of length(word[i]) * length(word[j]) where the two words do not share common letters. If no such two words exist, return 0.

 

Example 1:

Input: words = ["abcw","baz","foo","bar","xtfn","abcdef"]
Output: 16
Explanation: The two words can be "abcw", "xtfn".
Example 2:

Input: words = ["a","ab","abc","d","cd","bcd","abcd"]
Output: 4
Explanation: The two words can be "ab", "cd".
Example 3:

Input: words = ["a","aa","aaa","aaaa"]
Output: 0
Explanation: No such pair of words.
 

Constraints:

2 <= words.length <= 1000
1 <= words[i].length <= 1000
words[i] consists only of lowercase English letters.


To solve this problem efficiently, we can use **bit masking**. The idea is:

- Each word consists of lowercase letters (`'a'` to `'z'`), so we can represent each word using a **26-bit integer**, where each bit represents whether a particular letter exists in the word.
- This way, we can **quickly check if two words share common letters** using bitwise AND (`&`). If `mask[i] & mask[j] == 0`, the words do **not** share any common characters.

### ðŸ”§ Steps:
1. Precompute bitmasks for each word.
2. Compare each pair of words:
   - If they do not share common letters, compute the product of their lengths.
   - Keep track of the maximum product.

### âœ… Python Code:
```python
def maxProduct(words):
    n = len(words)
    masks = [0] * n
    lengths = [len(word) for word in words]

    # Create bitmask for each word
    for i in range(n):
        for ch in words[i]:
            masks[i] |= 1 << (ord(ch) - ord('a'))

    max_product = 0

    # Compare all pairs
    for i in range(n):
        for j in range(i + 1, n):
            if masks[i] & masks[j] == 0:  # No common letters
                max_product = max(max_product, lengths[i] * lengths[j])

    return max_product
```

```python
print(maxProduct(["a","aa","aaa","aaaa"]))                          # Output: 0
```

```python
def max_word_product(words):
```
